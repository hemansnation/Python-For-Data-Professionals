# Handling missing values

```python
df = pd.read_csv('weather_data2.csv', parse_dates=['day'])
```

```python
df
```

```python
import numpy as np
np.nan
```

```python
print(type(df.day[0]))
```

```python
df.set_index('day', inplace=True)
df
```

## fillna

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'No Event'
})
new_df
```

## forward fill

```python
new_df = df.fillna(method='ffill')
new_df
```

## backward fill

```python
new_df = df.fillna(method='bfill')
new_df
```

## interpolate

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method='time')
new_df
```

## dropna

```python
df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```
